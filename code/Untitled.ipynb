```python
#import mods
import requests
from bs4 import BeautifulSoup
```

```python
#Request page from the specified url
page = requests.get('https://en.wikipedia.org/wiki/Weightlifting_at_the_1976_Summer_Olympics')
#check the response's <status code> and <content>
print(page.status_code)
print(page.content)

#scrape webpage with BeautifulSoup <3
soup = BeautifulSoup(page.contents, 'html.parser')
#display the scrapped data
print(soup.prettify())
```
